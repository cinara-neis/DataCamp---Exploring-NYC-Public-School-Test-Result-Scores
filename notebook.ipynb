```python
best_math_schools = (schools[["school_name","average_math"]].loc[schools.average_math > 640]).sort_values("average_math",ascending=False)
best_math_schools
```

```python
schools["total_SAT"] = schools[["average_math", "average_reading", "average_writing"]].sum(axis=1)
```

```python
top_10_schools = schools[["school_name", "total_SAT"]].nlargest(10, "total_SAT")
```

```python
schools['num_schools'] = schools.groupby('borough')['borough'].transform('count')
```

```python
schools['average_SAT'] = schools.groupby('total_SAT')['total_SAT'].transform('mean').round(2)
```

```python
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2).reset_index()

largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()].set_index("borough")

largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})

```
